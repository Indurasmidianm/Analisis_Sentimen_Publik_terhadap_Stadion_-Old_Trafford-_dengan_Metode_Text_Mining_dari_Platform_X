# Data Cleaning & Preprocessing

```python
def cleaningText(text):
    text = re.sub(r'@[A-Za-z0-9]+', '', text)
    text = re.sub(r'#[A-Za-z0-9]+', '', text)
    text = re.sub(r'RT[\s]', '', text)
    text = re.sub(r"http\S+", '', text)
    text = re.sub(r'[0-9]+', '', text)

    text = text.replace('\n', ' ')
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.strip(' ')
    return text

def casefoldingText(text):
    text = text.lower()
    return text

def removeSlangWords(text):
    slang_words = slang_dict
    words = text.split()
    words = [slang_words[word] if word in slang_words else word for word in words]
    text = ' '.join(words)
    return text

    return ' '.join(words)
def tokenizingText(text):
    text = word_tokenize(text)
    return text

def filteringText(text):
    listStopwords = set(stopwords.words('english')) #inggris
    filtered = []
    for txt in text:
        if txt not in listStopwords:
            filtered.append(txt)
    text = filtered
    return text

def stemmingText(text):
    stemmer = PorterStemmer()
    text = [stemmer.stem(word) for word in text]
    return text

def toSentence(list_words):
    sentence = ' '.join(word for word in list_words)
    return sentence

def stemming(tokens):
    return " ".join(tokens)   # ubah list jadi string

def plot_cloud(wordcloud):
    plt.figure(figsize=(20,10))
    plt.imshow(wordcloud)
    plt.axis('off')

def removeSpecialWords(text):

    special_words = [
        "old",
        "trafford",
        "old trafford",
        "israel",
        "manchester",
        "united",
        "match",
        "player"
    ]

    for word in special_words:
        text = re.sub(r'\b' + re.escape(word) + r'\b', '', text)

    return text
```

```python
df['text_clean'] = df['full_text'].apply(cleaningText)
df['text_clean'] = df['text_clean'].apply(removeSpecialWords)
df['text_clean'] = df['text_clean'].apply(casefoldingText)
df['text_clean'] = df['text_clean'].apply(removeSlangWords)
df['text_clean'] = df['text_clean'].apply(tokenizingText)
df['text_clean'] = df['text_clean'].apply(filteringText)
df['text_clean'] = df['text_clean'].apply(stemming)
df['text_clean'] = df['text_clean'].apply(tokenizingText)
```

```python
df["text_vader"] = df["text_clean"].apply(toSentence)
```

# Sentiment Polarity  with English Sentiment Lexicon Vader
###### Credit: Hutto, C.J. & Gilbert, E.E. (2014). VADER: A Parsimonious Rule-based Model for Sentiment Analysis of Social Media Text. Eighth International Conference on Weblogs and Social Media (ICWSM-14). Ann Arbor, MI, June 2014.

```python
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sia = SentimentIntensityAnalyzer()
```

```python
def vader_sentiment(text):
    score = sia.polarity_scores(text)
    compound = score['compound']

    if compound >= 0.05:
        return compound, "positive"
    elif compound <= -0.05:
        return compound, "negative"
    else:
        return compound, "neutral"
```

```python
results = df['text_vader'].apply(vader_sentiment)
results = list(zip(*results))
df['polarity_score'] = results[0]
df['polarity'] = results[1]
print(df['polarity'].value_counts())
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
counts = df['polarity'].value_counts()
labels = list(counts.index)
sizes = list(counts.values)
explode = (0.1, 0, 0)
ax.pie(x=sizes, labels=labels, autopct='%1.1f%%', explode=explode, textprops={'fontsize': 14})
ax.set_title('Sentiment Polarity', fontsize=16, pad=20)

total = sum(sizes)

plt.show()
```

## Top 2 - Grams

```python
from nltk import ngrams
from collections import Counter
import plotly.express as px
import pandas as pd

# Distribusi Frekuensi N-gram
n = 2

ngram_counts = df['text_clean'].apply(lambda x: list(ngrams(x, n)))
ngram_counts = [ngram for ngrams_list in ngram_counts for ngram in ngrams_list]

ngram_freq = Counter(ngram_counts)

labels, values = zip(*ngram_freq.most_common(10))
labels = [' '.join(label) for label in labels]

# Create dataframe for plotly
ngram_df = pd.DataFrame({
    'N-gram': labels,
    'Frequency': values
})

# Interactive bar chart
fig = px.bar(
    ngram_df,
    x='N-gram',
    y='Frequency',
    title=f'Top 10 Most Common {n}-grams',
    text='Frequency'
)

fig.update_layout(
    xaxis_title='N-gram',
    yaxis_title='Frequency',
    xaxis_tickangle=-45
)

fig.show()
```
